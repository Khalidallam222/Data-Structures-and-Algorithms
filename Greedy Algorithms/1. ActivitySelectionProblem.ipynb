# The Greedy Approach:
1. Sort times by end time
2. Select the first activity
3. For each remaining activity, if the start time is greater than or equal to the end time of the previously selected activity, select it

````python
def printMaxActivities(s, f):
    n = len(f)
    print("The following activities are selected: ")

    # The first activity is always selected
    i = 0
    print(i, end = ' ')

    # For each remaining activity, if the start time is greater than or equal to the end time 
    # of the previously selected activity, select it otherwise skip it
    j = 0
    for i in range(1, n):
        if s[i] > f[j]:
            print(i, end = ' ')
            j = i

# Driver code
s = [1, 3, 0, 5, 8, 5]
f = [2, 4, 6, 7, 9, 9]

# Function call
printMaxActivities(s, f)
````

- Time Complexity: O(nlogn) for sorting and selecting activities and O(n) for selecting activities only
- space complexity: O(n) for sorting and selecting activities and O(1) for selecting activities only

# How to implement when given activities are not sorted?

```f1 = lambda x : x[1]
Activity = [[5, 9], [1, 2], [3, 4], [0, 6], [5, 7], [8, 9]]
Activity.sort(key=f1)
print(Activity)
```

1. first it will apply the function f1 on each element of the list
2. then it will sort the list based on the result of the function f1
3. finally it will return the sorted list

````python
def MaxActivities(Activity):
    selected = []
    
    # Sort jobs according to their finish time
    Activity.sort(key = lambda x: x[1])
    selected.append(Activity[0])
    j = 0
    for i in range(1, len(Activity)):
        if Activity[i][0] > Activity[j][1]:
            selected.append(Activity[i])
            j = i
    return selected


#Driver code 
Activity = [[5, 9], [1, 2], [3, 4], [0, 6], [5, 7], [8, 9]]
selected = MaxActivities(Activity)
print("Following activities are selected :")
print(selected[0], end = "")
for i in range (1, len(selected)):
    print(",", end = " ")
    print(selected[i], end = '')
    
````
